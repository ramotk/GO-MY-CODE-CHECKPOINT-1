```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as mpl

a=pd.read_csv("titanic-passengers.csv",delimiter=';')
a
```

```python
print(a.isnull().sum())
```

```python
a['Age'].fillna( a["Age"].mean())
a
```

```python
a.drop('Cabin',axis = 1 ,inplace = True)
```

```python
td = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess": "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
```

```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = mpl.subplots( figsize =( 17 , 17))
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={'shrink' : .9}, 
        ax=ax, 
        annot = True, 
        annot_kws = {'fontsize' : 12})

    
plot_correlation_map(a)
```
