```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as mpl

a=pd.read_csv("titanic-passengers.csv",delimiter=';')
a
```

```python
print(a.isnull().sum())
```

```python
a['Age'].fillna( a["Age"].mean(), inplace=True)
a['Survived'].replace({'No':0, 'Yes' : 1}, inplace=True)
a['Sex']=a['Sex'].map({'male':1, 'female' : 0})
a['Embarked'].fillna('S', inplace=True)
a['Embarked'].replace({'Q':2 ,'S':1, 'C':0}, inplace =True)
a.head()
```

```python
print (a.isnull().sum())
```

```python
a.drop('Ticket', axis = 1, inplace=True)
a.drop('Cabin',axis = 1 ,inplace = True)
```

```python
td = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess": "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
```

```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = mpl.subplots( figsize =( 17 , 17))
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={'shrink' : .9}, 
        ax=ax, 
        annot = True, 
        annot_kws = {'fontsize' : 12})

    
plot_correlation_map(a)
```

```python
noms = a['Name'].values
a['Title']= list(map(lambda x : x.split(",")[1].split('.')[0].strip(), noms))
a['Title'].replace(td, inplace=True)
a['Title'].values
v = {td:num for num,td in enumerate(list(td.values()))}
for x in v : v[x] = range(len(v))[list(v.keys()).index(x)]
a["Title"].replace(v, inplace = True)
a.drop('Name',axis = 1 ,inplace = True)
a.head()
```

```python
a['Fam'] = (a['Parch'] + a['SibSp'])
a['Fam']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
r = list(a.columns)
r.remove("Survived")
x = a[r]
y = a['Survived']

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25, random_state= 0)
```

```python
logreg = LogisticRegression()   
logreg.fit(x_train, y_train)  
y_pred  = logreg.predict(x_test)    
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```
